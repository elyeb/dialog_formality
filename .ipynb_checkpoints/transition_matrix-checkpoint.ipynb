```python
import datatable
import json
from tqdm import tqdm
import matplotlib.pyplot as plt
```

```python
with open('../CRECIL/Final_Data/train.json','rb') as infile:
    train_df = json.loads(infile.read())
```

```python
# 1. Get co-occurence counts
rel_counts = dict()
all_rels = set()

for i in tqdm(range(0,len(train_df))):
    for j in range(0,len(train_df[i][1])):
        rel_list = train_df[i][1][j]['r'] #list of variable length with relation ids (0-31)
        for r in rel_list:
            all_rels.add(r)
            if r not in rel_counts:
                rel_counts[r] = dict()

                
            for k in range(0,len(train_df[i][1])):
                if k!=j:
                    other_rels = train_df[i][1][k]['r']
                    for other_r in other_rels:
                        if other_r not in rel_counts[r]:
                            rel_counts[r][other_r] = 0
                        rel_counts[r][other_r] += 1
```

```python
#get divisor
r_dist = dict() #for all relation counts, future divisor
for r in all_rels:
    r_dist[r] = 0
for i in range(0,len(all_rels)):
    for j in range(0,len(all_rels)):
        
        r_dist[all_rels[j]] += rel_counts[all_rels[i]][all_rels[j]]
        
 
```

```python
# 3. Convert to probabilities - if you do this right it should all sum to 1
rel_probs = dict()
for i in range(0,len(all_rels)):
    rel_probs[all_rels[i]] = dict()
    
    for j in range(0,len(all_rels)):
        rel_probs[all_rels[i]][all_rels[j]] = rel_counts[all_rels[i]][all_rels[j]]/r_dist[all_rels[j]]
```
